```python
class member:
    def __init__(self,name,membershipID):
        self.name = name
        self.membershipID = membershipID

    def display(self):
        print("Name:", self.name)
        print("Membership ID:",self.membershipID)


class RegularMember(member):
    def _init_(self, name, membershipID):
        super()._init_(name, membershipID)
        self.borrow_limit = 3

    def getBorrowLimit(self):
         return self.borrow_limit

    def display(self):
        print("\nMember Type: Regular Member")
        super().display()
        print("Borrow Limit:", self.getBorrowLimit())


class PremiumMember(member):
    def _init_(self, name, membershipID):
        super()._init_(name, membershipID)
        self.borrow_limit = 10
    def getBorrowLimit(self):
        return self.borrow_limit

    def display(self):
        print("\nMember Type: Premium Member")
        super().display()
        print("Borrow Limit:", self.getBorrowLimit())

def main():
    regular = RegularMember("ranjani", 101)
    premium = PremiumMember("priya", 102)

    regular.display()
    premium.display()

if __name__=="__main__":
     main()



```
